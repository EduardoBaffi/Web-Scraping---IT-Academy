##### BeautifulSoup

```python
import requests
from bs4 import BeautifulSoup

URL = "https://www.bolsamadrid.es"
requests.get(URL)
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
# Prettify method: print out the HTML content of the page formatted nicely.
print(soup.prettify())
```

```python
# The objective is to do Web Scrapping about companies listed on the Madrid Stock Exchange.
# Search the available links of the HTML showed above

links = soup.find_all('a')
for page in links:
    print(page)
```

```python
# Search the links of listed companies ( "Lista de Empresas")
links = soup.find_all('a', string = 'Lista de Empresas')
for page in links:
    print(page)
```

```python
# 'href' property of the tag determines where the link goes.
# Now, we checking the code with the help of www.freeformatter.com to identify the 'id' of the source of data
# The necessary dat is avaible here: 
# <div>
#        <table class="TblPort" cellspacing="0" cellpadding="3" id="ctl00_Contenido_tblEmisoras" style="width:100%;border-collapse:collapse;">

results = soup.find(id="ctl00_Contenido_tblEmisoras")
```

```python
import pandas as pd

# Define the link of the table
URL_new = 'https://www.bolsamadrid.es/esp/aspx/Empresas/Empresas.aspx'
page_new = requests.get(URL_new)
soup_table = BeautifulSoup(page_new.content, "html.parser")

table_data = soup_table.find('table', class_="TblPort")

# Headers of the table
headers = []
for i in table_data.find_all('th'):
    title = i.text
    headers.append(title)

    #Body of the table
df = pd.DataFrame(columns = headers)

for j in table_data.find_all('tr')[1:]:
        row_data = j.find_all('td')
        row = [tr.text for tr in row_data]
        length = len(df)
        df.loc[length] = row
```

```python
# Web Scraped Table:
df
```

```python
# Stocks Table - 26/7/2021

import requests
import pandas as pd

# Define the link of the table of the englich version of the webpage
URL_stocks = 'https://www.bolsamadrid.es/ing/aspx/Mercados/Precios.aspx?indice=ESI100000000'
page_stocks = requests.get(URL_stocks)
soup_stocks = BeautifulSoup(page_stocks.content, "html.parser")

table_data = soup_stocks.find(id="ctl00_Contenido_tblAcciones")

# Headers of the table
headers = []
for i in table_data.find_all('th'):
    title = i.text
    headers.append(title)

    #Body of the table
df_stocks = pd.DataFrame(columns = headers)

for j in table_data.find_all('tr')[1:]:
        row_data = j.find_all('td')
        row = [tr.text for tr in row_data]
        length = len(df_stocks)
        df_stocks.loc[length] = row
```

```python
# Export table to csv
df_stocks.to_csv('stocksBolsaMadrid.csv')
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

options = Options()
options.headless = True
options.add_argument("--window-size=1920,1200")

DRIVER_PATH= r'C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe'

driver = webdriver.Chrome(options=options, executable_path=DRIVER_PATH)
driver.get("https://www.bolsamadrid.es")
print(driver.page_source)
#driver.quit()
```

```python
web = driver.find_elements_by_link_text('Acciones')
```

```python
print(web)
```

```python
# There two links, we choose the first one
web[0].click()
```

```python
# After the click, we enter the next tab
# The table is the new tab. Let`s get its data

data = driver.find_element_by_id('ctl00_Contenido_tblAcciones')
```

```python
# Headers of the table
headers = []
for element in data.find_elements_by_tag_name('th'):
    title = element.text
    headers.append(title)
print(headers)
```

```python
row = []

row_data = data.find_elements_by_tag_name('td')
row = [element.text for element in row_data]
```

```python
print(row)
```

```python
# The table have nine columns. 
# We will create a dictionary Because of that em salto cada 9 elements per tenir els mateix tipus de valors

values = {}
for j in range(9):
    values[headers[j]]=row[j::9]
```

```python
#Create Dataframe

pd.DataFrame(values)
```
